```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
with open("Raw/I94_SAS_Labels_Descriptions.SAS") as f:
    contents = f.readlines()
```

```python
i94cntyl = contents[9:298]

countries = pd.DataFrame(i94cntyl, columns=['column_name'])
countries.replace('\n', '', regex=True, inplace=True)
countries.replace("'", '', regex=True, inplace=True)

countries[['country_code','country']] = countries['column_name'].str.split('=',expand=True)
countries = countries[['country_code','country']]

# filter out 'no country code' and 'invalid'


countries['country_code'] = countries['country_code'].astype(int)

country_residence = countries.rename(columns = {'country_code' : 'country_residence_code', 
                                                 'country' : 'country_residence'})

country_citizenship = countries.rename(columns = {'country_code' : 'country_citizenship_code', 
                                                 'country' : 'country_citizenship'})

countries

```

```python
i94prtl = contents[302:962]
airports = pd.DataFrame(i94prtl, columns=['column_name'])
airports.replace('\n', '', regex=True, inplace=True)
airports.replace("'", '', regex=True, inplace=True)
airports.replace('\t', '', regex=True, inplace=True)
airports[['arrival_airport_code','arrival_airport']] = airports['column_name'].str.split('=',expand=True)
airports = airports[['arrival_airport_code','arrival_airport']]

airports['arrival_airport_code'] = airports['arrival_airport_code'].astype(str)

airports

# filter out no port code
```

```python
i94model = contents[972:976]
modes = pd.DataFrame(i94model, columns=['column_name'])
modes.replace('\n', '', regex=True, inplace=True)
modes.replace("'", '', regex=True, inplace=True)
modes.replace('\t', '', regex=True, inplace=True)
modes[['mode_entry_code','mode_entry']] = modes['column_name'].str.split('=',expand=True)
modes = modes[['mode_entry_code','mode_entry']]


modes['mode_entry_code'] = modes['mode_entry_code'].astype(int)

modes

# filter out not reported
```

```python
i94addrl = contents[981:1036]
US_states = pd.DataFrame(i94addrl, columns=['column_name'])
US_states.replace('\n', '', regex=True, inplace=True)
US_states.replace("'", '', regex=True, inplace=True)
US_states.replace('\t', '', regex=True, inplace=True)
US_states[['arrival_state_code','arrival_state']] = US_states['column_name'].str.split('=',expand=True)
US_states = US_states[['arrival_state_code','arrival_state']]
US_states

# filter out all other codes
```

```python
i94visa = contents[1046:1049]
visa_codes = pd.DataFrame(i94visa, columns=['column_name'])
visa_codes.replace('\n', '', regex=True, inplace=True)
visa_codes.replace("'", '', regex=True, inplace=True)
visa_codes.replace('\t', '', regex=True, inplace=True)
visa_codes[['visa_code','visa_category']] = visa_codes['column_name'].str.split('=',expand=True)
visa_codes = visa_codes[['visa_code','visa_category']]
visa_codes['visa_code'] = visa_codes['visa_code'].astype(int)
visa_codes
```
